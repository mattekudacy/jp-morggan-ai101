```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
plt.show()
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Demo Plot')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

```python
figure = plt.figure()

axes1 = figure.add_axes([0.1, 0.1, 0.8, 0.8]) #main axes
axes2 = figure.add_axes([0.2, 0.5, 0.4, 0.3]) #insert axes 

# larger figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('Set X Label')
axes1.set_ylabel('Set Y Label')
axes1.set_title('Axes 2 Title')

# larger figure Axes 2
axes2.plot(x, y, 'r')
axes2.set_xlabel('Set X Label')
axes2.set_ylabel('Set Y Label')
axes2.set_title('Axes 2 Title')
```

```python
# Using Subplots
figure , axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Axes 2 Title')
```

```python
# Empty Canvas
figure, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
figure

```

```python
# fig.tight_layout() or plot.tight_layout() method
figure, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
figure
# plt.tight_layout()
figure.tight_layout()
```

```python
figure, axes = plt.subplots(figsize=(12, 3))
axes.plot(x, y, 'g')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
# Legends
figure = plt.figure()

ax = figure.add_axes([0, 0, 1, 1]) #main axes

# larger figure Axes 1
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
figure = plt.figure()

ax = figure.add_axes([0, 0, 1, 1]) #main axes

# larger figure Axes 1
ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
ax.legend()
```
